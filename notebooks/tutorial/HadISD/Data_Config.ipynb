### Configure The Download Directory
By default, HadISD data data will be downloaded to a directory named "HadISD_data" in the user's home directory.

Users can change the directory where data is downloaded and converted to Zarr by changing the `default_dir` variable, or by setting the `HADISD_DOWNLOAD_DIR` environment variable.

```python
import os
from pathlib import Path

# Set the download directory (user can change this if desired)
default_dir = Path.home() / "HadISD_data"

download_dir = os.environ.get("HADISD_DOWNLOAD_DIR", str(default_dir))
download_dir = Path(download_dir)
download_dir.mkdir(parents=True, exist_ok=True)
```
